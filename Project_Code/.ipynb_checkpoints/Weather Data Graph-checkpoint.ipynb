```python
import pandas
import matplotlib.pyplot as plt
import numpy as np
import datetime

weatherdata = pandas.read_csv('test.csv')

TIMES = pandas.to_datetime(weatherdata.CURRENT_TIME)
plt.plot_date(TIMES, weatherdata.TEMP, linestyle='solid')

plt.gcf().autofmt_xdate()

plt.show()

```
